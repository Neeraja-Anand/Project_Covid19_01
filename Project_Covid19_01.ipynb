```python
confirmedcases = df.groupby('Country/Region')['Confirmed'].sum()

total_confirmedcases = confirmedcases.sort_values(ascending=False).head(5).reset_index()

print(total_confirmedcases)
```

```python
x_axis = "Confirmed"
y_axis = "Country/Region"
plt.figure(figsize=(9, 6))
fig=sns.barplot(x='Confirmed', y='Country/Region', data=total_confirmedcases, color='Red')
plt.xlabel("Confirmed Cases", fontsize=10)
plt.ylabel("Countries", fontsize=10)
plt.title("Top 20 Countries by Total Cases", fontsize=12)
plt.show()
```

```python
Death = df.groupby("Country/Region")["Deaths"].sum()
deaths_rate = Death.sort_values(ascending=False).head(10).reset_index()
print(deaths_rate)
```

```python
x_axis = "Deaths"
y_axis = "Country/Region"

plt.figure(figsize=(9, 6))
sns.barplot(x='Deaths', y='Country/Region', data=deaths_rate, color='Red')
plt.xlabel("Total Deaths", fontsize=15)
plt.ylabel("Countries", fontsize=15)
plt.title("Top 5 Countries by Total Deaths", fontsize=10)
plt.show()
```

```python
region = df.groupby("WHO Region")[["Confirmed", "Deaths"]].sum()
region = region.sort_values(by=["Confirmed", "Deaths"], ascending=False).reset_index()
print(region)
```
